```python
import matplotlib.pyplot as plt
x = range(10+1)
plt.plot(x)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.show()
```

```python
import numpy as np
x = np.array(range(10+1))
y = x**2
# rgb + 0
plt.plot(x,y,'bo')
plt.xlabel('x label')
plt.ylabel('y label')
plt.show()
```

```python
import numpy as np
x = np.array(range(10+1))
y = x**2
# rgb + o-
plt.plot(x,y,'bo-')
plt.xlabel('x label')
plt.ylabel('y label')
plt.show()
```

```python
import numpy as np
x = np.array(range(10+1))
y = x**2
# rgb + o-
plt.plot(x,y,'rs--')
plt.xlabel('x label')
plt.ylabel('y label')
plt.show()
```

```python
import numpy as np
x = np.array(range(10+1))
y = x**2
# rgb + o-
plt.plot(x,y,'rs--')
plt.xlabel('x label')
plt.ylabel('y label')
plt.axis([-5,10,-20,110])
plt.show()
```

```python
t = np.arange(0,10,0.5)
plt.plot(t,t,'ro-',t,t**2, 'bs-', t,t**3, 'g--')
plt.show()
```

```python
t = np.arange(0,10,0.5)
plt.plot(t,t,'ro-',t,t**2, 'bs-', t,t**3, 'g--', t, t**4, 'go',linewidth=3)
plt.show()
```

```python
import pandas as pd
randomSequence = pd.DataFrame(np.random.normal(1.0, 0.08, (100,8)))
accunulate = randomSequence.cumprod()
accunulate.plot()
plt.title('behavior')
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend(loc=0)
plt.show()
```
